```python
n=input()
i=len(n)-1

reverse=""

print(i)

while(i>=0):
    reverse+=n[i]
    i-=1
    
  
    
if reverse==n:
        print("its a palindrom")
        
else:
    print("its not a palindrome")


print("reversed: ",reverse)

```
